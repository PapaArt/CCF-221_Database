```python
import sqlite3
import pandas as pd

db = sqlite3.connect('data/hubway.db')

def run_query(query):
    return pd.read_sql_query(query, db)
```

```python
query = 'SELECT start_date, bike_number FROM trips LIMIT 5;'

run_query(query)
```

```python
query = 'SELECT duration, start_date FROM trips LIMIT 5'

run_query(query)
```

```python
query = '''

SELECT duration FROM trips 

ORDER BY duration DESC 

LIMIT 1;

'''

run_query(query)
```

```python
query = '''

SELECT duration FROM trips 

ORDER BY duration DESC 

LIMIT 10;

'''

run_query(query)
```

```python
query = '''

SELECT * FROM trips

WHERE bike_number = "B00400"

LIMIT 5;

'''

run_query(query)
```

```python
query = '''

SELECT * FROM trips

WHERE duration > 9990;

'''

run_query(query)
```

```python
query = '''

SELECT * FROM trips

WHERE (duration >= 9990) AND (sub_type = "Registered")

ORDER BY duration DESC;

'''

run_query(query)
```

```python
query = '''

SELECT COUNT(*) FROM trips

WHERE sub_type = "Registered";

'''

run_query(query)
```

```python
query = '''

SELECT COUNT(*) AS "Total Trips by Registered Users"

FROM trips

WHERE sub_type = "Registered";

'''

run_query(query)
```

```python
query = '''

SELECT AVG(duration) AS "Average Duration"

FROM trips;

'''

run_query(query)
```

```python
query = '''

SELECT sub_type, AVG(duration) AS "Average Duration"

FROM trips

GROUP BY sub_type;

'''

run_query(query)
```
